````python
import json
with open('compiled_courses.json', 'r') as j:
    data = json.load(j)
````

````python
# Remove non relivant examples

for key, item in d.copy().items():

    num = key.split(' ')[-1].lower()

    if not num.isnumeric():

        d.pop(key)
    
    elif int(num) > 600:

        d.pop(key)
````

````python
from collections.abc import Callable
from typing import List

def search(func: Callable) -> List:

    return [li for li in d.values() if func(li)]
````

````python
d['CAS BI 203']
````

````python
search(lambda li: li['code'][:6] == 'CAS BI' and (li['attributes'] and 'PLM' in [a['value'] for a in li['attributes'] if a['code'] == 'HUB']))
````

````python
cs = search(lambda li: li['units'] < 3 and li['code'][:6] in ['CDS DS', 'CAS BI'])
````

````python
print(
    *[f'{csi['code']}' for csi in cs],
    sep='\n'
)
````
